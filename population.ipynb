```python
popestimate_cols = pop_2019.filter(like='POPESTIMATE').columns
pop_2019 = pd.concat([pop_2019['NAME'], pop_2019[popestimate_cols]], axis=1)
pop_2019 = pop_2019.iloc[5:56,:]
pop_2019.drop([13], inplace=True)
```

```python
pop_2019 = pop_2019.reset_index(drop=True)
```

```python
pop_2019_df = pop_2019.melt(id_vars=['NAME'], var_name='year', value_name='population')
pop_2019_df['year']
```

```python
pop_2019_df.rename(columns={'NAME': 'state'}, inplace=True)
pop_2019_df = pop_2019_df[pop_2019_df['year'] > 2010]
```
