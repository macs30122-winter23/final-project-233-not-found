```python
import re
import pandas as pd
import csv
```

```python
ab_data = pd.read_csv(r"data\abortion_data\abortion data.csv")
ab_data.drop(ab_data[(ab_data['state'] == 'New York City')].index, inplace=True)
ab_data.drop(ab_data[(ab_data['state'] == 'New York State')].index, inplace=True)
ab_data = ab_data.reset_index(drop=True)
ab_data
```

```python
# clean state names && numerical data
states = ab_data['state'].replace(r'¶|§|†|\*', '' , regex=True).to_frame()
occur_by_year = ab_data.iloc[:, 1:].apply(lambda x: x.str.replace(',', "").astype(int), axis=1).reset_index(drop=True)
ab_df = pd.concat([states, occur_by_year], axis=1)
ab_df
```

```python
reconstruct_df = ab_df.melt(id_vars=['state'], var_name='year', value_name='abortion_occurrences')
reconstruct_df
```

```python
reconstruct_df.to_csv("ab_occurrence.csv")
```
