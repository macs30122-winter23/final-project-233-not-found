```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Abortion_law_in_the_United_States_by_state"
upper_url = "https://en.wikipedia.org"
```

```python
state_names = ["Alaska", "Alabama", "Arkansas",  "Arizona", 
"California", "Colorado", "Connecticut", "Delaware", 
"Florida", "Georgia", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", 
"Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", 
"Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota",
 "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", 
 "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", 
 "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", 
 "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
len(state_names)
```

```python
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
links = []

for state in state_names:
    if state == "Washington":
        links.append(upper_url+ "/wiki/Abortion_in_" + state +"_(state)")  # otherwise confusion of state & DC
    else:
        links.append(upper_url + soup.find('a', href=re.compile("/wiki/Abortion_in_" + re.sub('\s', '_', state))).get('href'))
print(links)
```

```python
''' get a list of all states abortion law legal or not '''
# i = 1
legal_or_not = []
for link in links:
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    txt = soup.find('div', id='mw-content-text').find('p',class_=None).text  # some pages with different layout
    if txt == "":
        txt = soup.find(class_='mw-content-container').find('p').text   # majority of wiki pages
    # print(i, link)
    # print(txt)
    legal_or_not.append(re.search('legal|illegal', txt).group())
    # i+=1
```

```python
''' create the df '''
abortion_df = pd.DataFrame({"state": state_names, "legal?": legal_or_not})
abortion_df['legal?'] = abortion_df['legal?'].replace({'legal': 1, 'illegal': 0})
abortion_df
```
