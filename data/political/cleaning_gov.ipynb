```python
import pandas as pd
import os
import glob
import re
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
dfs = []
# loop over the list of csv files
for f in csv_files:
    # filename = f.split("\\")[-1]
    # year = re.findall(r"\d+", filename)[-1]
    # if year in ['2015','2016','2017','2018', '2019', '2020']:
    # read the csv file
    df = pd.read_csv(f)
    dfs.append(df)
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
df = pd.concat(dfs, ignore_index=True)
```

```python
df.head()
```

```python
# select interested columns
df = df.iloc[:, [1, 7, 8]]
df.shape
```

```python
# one-hot-encoding
dummies = pd.get_dummies(df[['Position']])
df_gov = pd.concat([df, dummies], axis=1)
df_gov = df_gov.drop(columns=['Position'])
```

```python
# clean the state name
df_gov['state'] = df_gov['state'].str.findall(r'[A-Z]{2}').str[0]
```

```python
# convert year to integer
df_gov['Year'] = df_gov['Year'].astype(object)
```

```python
df_leg = pd.read_csv('/Users/axyz1220/Downloads/macs30122/final_project/legislature/women_in_legislature.csv')
```

```python
df_poli = pd.merge(df_leg, df_gov, how='left', left_on=['year', 'state'], right_on = ['Year', 'state'])
```

```python
df_poli
```

```python
# only keep one column of year
df_poli = df_poli.drop(columns=['Year'])
```

```python
# fill na with 0, indicating male governor or lt_governor
df_poli = df_poli.fillna(0)
```

```python
def convert_to_full_name(abb):
    """
    Convert abbreviation to full state name

    Inputs:
        abb: str
    
    return:
        full state name
    """
    return states_full_name[abb]
```
