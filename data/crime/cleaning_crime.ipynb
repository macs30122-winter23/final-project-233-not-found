```python
import pandas as pd
```

```python
df_crime = pd.read_csv('state_crime_raw.csv')
df_crime.head()
```

```python
# select 2011~2019
df_crime = df_crime[df_crime['Year'] > 2010]
df_crime.shape
```

```python
df_crime.columns
```

```python
df_crime = df_crime[['State', 'Year', 'Data.Rates.Violent.Rape']]
df_crime.columns = ['state', 'year', 'rape_rate']
df_crime.shape
```

```python
state_names = ["Alaska", "Alabama", "Arkansas", "Arizona", "California", "Colorado", 
"Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Iowa", 
"Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", 
"Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", 
"New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", 
"Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Vermont", "Washington", 
"Wisconsin", "West Virginia", "Wyoming"]
```

```python
# only keep major 50 states
df_crime = df_crime[df_crime['state'].isin(state_names)]
```

```python
# import 2020 crime data
crime_2020 = pd.read_csv('rape_2020_raw.csv')
crime_2020.head()
```

```python
# add year column
crime_2020.columns = ['state', 'rape_rate']
crime_2020['year'] = 2020
```

```python
# drop headers
crime_2020 = crime_2020.drop([0, 1])
```

```python
# check null value
crime_2020.isnull()
```

```python
# select 50 major states
crime_2020 = crime_2020[crime_2020['state'].isin(state_names)]
crime_2020.shape
```

```python
# concatenate all years
df_crime = pd.concat([df_crime, crime_2020])
```

```python
df_crime.to_csv('rape_data.csv', index=False)
```
